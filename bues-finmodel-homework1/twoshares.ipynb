This notebook demonstrates how to download financial data in Python notebooks.

The next section sets up the environment.

```python
!pip -q install yfinance
!pip -q install matplotlib
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

```

The next section is dedicated to data download.

```python
def download_financial_data(ticker):
    data = yf.download(ticker, start="2023-01-01", end="2024-01-01")
    daily_returns = data['Adj Close'].pct_change()
    return daily_returns

apple_returns = download_financial_data('AAPL')
microsoft_returns = download_financial_data('MSFT')
```

The next section plots the two shares.

```python

plt.figure(figsize=(12,6))
plt.plot(apple_returns, label='Apple (AAPL)')
plt.plot(microsoft_returns, label='Microsoft (MSFT)')
plt.title('Daily Returns of AAPL and MSFT')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.legend()
plt.show()

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.hist(apple_returns, bins=50, alpha=0.7)
plt.title('Histogram of AAPL Returns')
plt.subplot(1,2,2)
plt.hist(microsoft_returns, bins=50, alpha=0.7)
plt.title('Histogram of MSFT Returns')
plt.tight_layout()
plt.show()

```

The next section calculates some stats.

```python
def display_distribution_indicators(returns, asset_name):
    print(f"Distribution Indicators for {asset_name}:")
    print(f"Mean: {returns.mean()}")
    print(f"Median: {returns.median()}")
    print(f"Standard Deviation: {returns.std()}")
    print(f"Kurtosis: {returns.kurtosis()}")
    print(f"Skewness: {returns.skew()}")

display_distribution_indicators(apple_returns, "Apple (AAPL)")
display_distribution_indicators(microsoft_returns, "Microsoft (MSFT)")
```
